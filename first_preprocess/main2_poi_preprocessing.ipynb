```python
dic = {}

for poi_na in pois.keys():
    #pos_id :: post_id
    for pos_id in pois[poi_na]['contents'].keys():
        dic[pos_id] = pois[poi_na]['name']
```

```python
f = open("data.txt", 'w')

#us_id :: user_id  
#po_id :: post_id
for us_id in users.keys():
    a=[]
    for po_id in users[us_id].keys():
        a.append((posts[po_id]['created_time'], dic[po_id], posts[po_id]['content']))

    a.sort()
    for (t, poi_names, contents) in a:
        f.write(us_id)
        f.write("\t")
        f.write(poi_names)
        f.write("\t")
        f.write(contents)
        f.write("\n")
    


f.close()
```
